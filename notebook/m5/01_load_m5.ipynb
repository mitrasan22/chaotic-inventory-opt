```python
import pandas as pd
sales=pd.read_csv("sales_train_validation.csv")
sales.head()
```

```python
item = "FOODS_3_090"
store = "CA_1"

sku = sales[
    (sales["item_id"] == item) &
    (sales["store_id"] == store)
]
sku.head()
```

```python
d_cols = [c for c in sku.columns if c.startswith("d_")]
demand = sku[d_cols].values.flatten().astype(float)
```

```python
demand
```

```python
import matplotlib.pyplot as plt
plt.plot(demand)
plt.title(f"Demand for {item}@{store}")
plt.show()
```
